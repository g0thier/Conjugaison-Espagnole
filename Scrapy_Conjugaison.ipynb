```python
# https://la-conjugaison.nouvelobs.com/espagnol/verbe/construir.php

liste_infinitif_url = []

for verbe in liste_infinitif:
    liste_infinitif_url.append("https://la-conjugaison.nouvelobs.com/espagnol/verbe/" + verbe + ".php")

liste_infinitif_url.sort()

print(liste_infinitif_url)
```

```python
import scrapy
from scrapy.crawler import CrawlerProcess

class QuotesSpider(scrapy.Spider):

    # Name of your spider
    name = "verbo"

    # Url to start your spider from 
    start_urls = liste_infinitif_url

    def parse(self, response):

        # //*[@id="contenu"]/div[6]/div # présent
        # //*[@id="contenu"]/div[7]/div # Passé composé
        # ...
        
        # //*[@id="contenu"]/div[6]/div/text()[1] # yo v...
        # //*[@id="contenu"]/div[6]/div/b[1] # ...uelvo
        
        
        indicatif = response.xpath('//*[@id="contenu"]/div[34]/div')
        quotes = response.xpath('//*[@id="contenu"]/div[6]/div')
        
        for quote in quotes:
            yield {
                
                'indicatif' : indicatif.xpath('text()[1]').get() + indicatif.xpath('b[1]/text()').get(),
                
                '1_pers_singulier' : quote.xpath('text()[1]').get() + quote.xpath('b[1]/text()').get(),
                '2_pers_singulier' : quote.xpath('text()[2]').get() + quote.xpath('b[2]/text()').get(),
                '3_pers_singulier' : quote.xpath('text()[3]').get() + quote.xpath('b[3]/text()').get(),
                
                '1_pers_pluriel' : quote.xpath('text()[4]').get() + quote.xpath('b[4]/text()').get(),
                '2_pers_pluriel' : quote.xpath('text()[5]').get() + quote.xpath('b[5]/text()').get(),
                '3_pers_pluriel' : quote.xpath('text()[6]').get() + quote.xpath('b[6]/text()').get(),
                
            }
```

```python
import logging

# Name of the file where the results will be saved
filename = "list_conjugaison.json"

## More info on built-in settings => https://docs.scrapy.org/en/latest/topics/settings.html?highlight=settings#settings
process = CrawlerProcess(settings = {
    'USER_AGENT': 'Chrome/97.0',
    'LOG_LEVEL': logging.ERROR,
    "FEEDS": {
        path + '/' + filename : {"format": "json"},
    }
})

# Start the crawling using the spider you defined above
process.crawl(QuotesSpider)
process.start()
```
